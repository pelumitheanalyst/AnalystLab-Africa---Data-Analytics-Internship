```python
# Imported seabornand numpy
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Create a boxplot of the log-transformed 'Quantity' column to reduce skewness and better visualize outliers
sns.boxplot(x=np.log1p(df["Quantity"]))
plt.xlabel("Quantity")
plt.title("Quantity Boxplot")
plt.show()
```

```python
# Create a boxplot of the log-transformed 'UnitPrice' column to reduce skewness and better visualize outliers
sns.boxplot(x=np.log1p(df["UnitPrice"]))
plt.xlabel("UnitPrice")
plt.title("UnitPrice Boxplot")

plt.show()
```

```python
# Calculate 25th percentile of Quantity
percentile25 = df['Quantity'].quantile(0.25)

#Calculate 75th percentile of Quantity
percentile75 = df['Quantity'].quantile(0.75)

#Calculate interquartile range
IQR = percentile75 - percentile25

#Calculate upper and lower limit for outliers
upper_limit = percentile75 + 1.5 * IQR
lower_limit = percentile25 - 1.5 * IQR

print('Lower limit is: ' + str(lower_limit))
print('Upper limit is: ' + str(upper_limit))
```

```python
# Calculate 25th percentile of UnitPrice
percentile25 = df['UnitPrice'].quantile(0.25)

#Calculate 75th percentile of UnitPrice
percentile75 = df['UnitPrice'].quantile(0.75)

#Calculate interquartile range
IQR = percentile75 - percentile25

#Calculate upper and lower limit for outliers
upper_limit = percentile75 + 1.5 * IQR
lower_limit = percentile25 - 1.5 * IQR

print('Lower limit is: ' + str(lower_limit))
print('Upper limit is: ' + str(upper_limit))
```

```python
# Identifed outliers in the 'Quantity' column using previously defined lower and upper limits
quantity_outliers = df[
    (df["Quantity"] < lower_limit) | 
    (df["Quantity"] > upper_limit)
]

# Created a scatter plot to visually display the detected outliers in Quantity
plt.figure(figsize=(10,4))
sns.scatterplot(x=quantity_outliers.index, y=quantity_outliers["Quantity"])
plt.title("Quantity Outliers")
plt.show()
```

```python
# Identifed outliers in the 'UnitPrice' column using previously defined lower and upper limits
price_outliers = df[
    (df["UnitPrice"] < lower_limit) | 
    (df["UnitPrice"] > upper_limit)
]

# Created a scatter plot to visually display the detected outliers in UnitPrice
plt.figure(figsize=(10,4))
sns.scatterplot(x=price_outliers.index, y=price_outliers["UnitPrice"])
plt.title("UnitPrice Outliers")
plt.show()
```

```python
# Displayed the top 20 most frequent StockCodes in the dataset
df['StockCode'].value_counts().head(20)
```

```python
# Identified StockCodes that do not follow the expected pattern of digits optionally followed by letters
df[~df['StockCode'].str.match(r'^\d+[A-Z]*$', na=False)]['StockCode'].unique()
```

```python
# Define a list of known invalid or non-product StockCodes and removed them from the dataset
invalid_codes = ['POST', 'C2', 'M', 'BANK CHARGES', 'PADS', 'DOT']

df = df[~df['StockCode'].isin(invalid_codes)]
```

```python
# Checked if any of the invalid StockCodes still exist in the dataset after filtering
df[df['StockCode'].isin(invalid_codes)].shape
```

```python
# Created a new column 'TotalSales' by multiplying Quantity and UnitPrice
df['TotalSales'] = df['Quantity'] * df['UnitPrice']
```

```python
# Calculated the median values of Quantity, UnitPrice, and TotalSales
df[['Quantity', 'UnitPrice', 'TotalSales']].median()
```

```python
# Calculated the total revenue generated across all transactions
df['TotalSales'].sum()
print(df['TotalSales'].sum())
```

```python
# Calculated total sales per invoice, and the average invoice value across all invoices
Average_order_value = df.groupby('InvoiceNo')['TotalSales'].sum().mean()
print(Average_order_value)
```

```python
# Identified the top 10 customers based on total sales generated
Top_customers = df.groupby('CustomerID')['TotalSales'].sum().sort_values(ascending=False).head(10)
print(Top_customers)
```

```python
# Identified the top 10 products based on total sales revenue
Top_products = df.groupby('Description')['TotalSales'].sum().sort_values(ascending=False).head(10)
print(Top_products)
```

```python
# Calculated the top 10 products by total revenue and visualized them using a bar chart
plt.figure(figsize=(12,6)) 

top_products = df.groupby('Description')['TotalSales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar')
plt.title("Top 10 Products by Revenue")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.savefig(r"C:\Users\adero\Downloads\New Projects\10_products_by_total_revenue.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
# Plotted daily sales trend to visualize how revenue changes over time
daily_sales = df.groupby(df['InvoiceDate'].dt.date)['TotalSales'].sum()

plt.figure(figsize=(12,5))
daily_sales.plot()

plt.title("Daily Sales Trend")
plt.ylabel("Revenue")

plt.savefig(r"C:\Users\adero\Downloads\New Projects\daily_sales_trend.png", dpi=300, bbox_inches="tight")
plt.show()
```
